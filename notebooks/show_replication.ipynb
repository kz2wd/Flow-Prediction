```python
import torch
import matplotlib.pyplot as plt

def plot_hist(target, name):
    # Plot histogram with matplotlib
    plt.figure(figsize=(6, 4))
    plt.hist(target, bins=200, range=(target.min(), target.max()), edgecolor='black')
    # plt.hist(target, bins=200, range=(-1, 1), edgecolor='black')
    plt.title(name)
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.grid(True)
    
    plt.show()
```

```python
from space_exploration.beans.dataset_bean import Dataset
from space_exploration.dataset import db_access
from space_exploration.dataset.transforms.AllTransforms import TransformationReferences
from space_exploration.models.AllModels import ModelReferences
from space_exploration.training import train_gan, get_split_datasets

model_ref = ModelReferences.C
dataset_name = "re200-sr005etot"

session = db_access.get_session()
dataset = Dataset.get_dataset_or_fail(session, dataset_name)


replication_ds = dataset.get_training_dataset(64, TransformationReferences.COMPONENT_NORMALIZE.transformation, TransformationReferences.Y_ALONG_COMPONENT_NORMALIZE.transformation, size=300)

x, y = next(iter(replication_ds))


plot_hist(y.cpu().numpy().reshape(-1), f'Histogram of y | {dataset_name}')

plot_hist(y[0].cpu().numpy().reshape(-1), f'Histogram of y 0 | {dataset_name}')
plot_hist(y[1].cpu().numpy().reshape(-1), f'Histogram of y 1 | {dataset_name}')
plot_hist(y[2].cpu().numpy().reshape(-1), f'Histogram of y 2 | {dataset_name}')

plot_hist(x.cpu().numpy().reshape(-1), f'Histogram of x | {dataset_name}')

plot_hist(x[0].cpu().numpy().reshape(-1), f'Histogram of x 0 | {dataset_name}')
plot_hist(x[1].cpu().numpy().reshape(-1), f'Histogram of x 1 | {dataset_name}')
plot_hist(x[2].cpu().numpy().reshape(-1), f'Histogram of x 2 | {dataset_name}')
```
